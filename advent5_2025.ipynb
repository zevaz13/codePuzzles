```python
def load_and_split(fileName):
    with open(fileName, 'r') as f:
        content = f.read().strip()

    # Split into two blocks
    block1, block2 = content.split("\n\n")

    # Convert each block into a list of lines
    rangeList = block1.splitlines()
    avList = block2.splitlines()
    return rangeList, avList

def rgList2ints(inList):
    starts,ends = [], []
    for r in inList:
        startx, endx = map(int, r.split("-"))
        starts.append(startx)
        ends.append(endx)
    return starts, ends

def available2ints(inList):
    avInt = []
    for i in inList:
        avx = int(i)
        avInt.append(avx)
    return avInt

def checkAvFresh(rangeFreshstarts, rangeFreshends, avVals):
    countFresh = 0
    freshElements = []
    for i in avVals:
        for s,e in zip(rangeFreshstarts, rangeFreshends):
            if i in range(s,e+1):
                countFresh += 1
                freshElements.append(i)
                print(f"val = {i} in [{s}, {e}]")
                break
    return countFresh, freshElements

### for part 2. Does not work with large datasets....
def checkUniqueFresh(rangeFreshstarts, rangeFreshends):
    allElements = []
    for si, ei in zip(rangeFreshstarts, rangeFreshends):
        for i in range(si,ei+1):
            allElements.append(i)
    UniqueFreshIDs = set(allElements)
    return UniqueFreshIDs

# we must not enumerate the elements of the intervals. 
def count_unique_from_ranges(starts, ends):
    # Pair and sort intervals
    intervals = sorted(zip(starts, ends)) # sorts intervals by start value

    merged = [] # to be filled with merged, non-overlapping intervals.
    cur_start, cur_end = intervals[0] # take the first start and end (remember, these are sorted)....

    for s, e in intervals[1:]:
        if s <= cur_end + 1:
            # Overlapping or touching → merge. Extends the current interval.
            cur_end = max(cur_end, e)
        else:
            # Disjoint → push previous interval
            merged.append((cur_start, cur_end))
            cur_start, cur_end = s, e

    merged.append((cur_start, cur_end))

    # Now compute total unique count. Only adds up the lengths of the merged intervals
    total = 0
    for s, e in merged:
        total += (e - s + 1)

    return total, merged
```
