```python
with open("day9_tst.txt", 'r') as f:
        content = f.read().split("\n")
content

inputLocs = []
for r in content:
    startx, endx = map(int, r.split(","))
    inputLocs.append([startx, endx])

#print(inputLocs)
```

```python
Loc1 = inputLocs[0]
Loc2 = inputLocs[1]

xdif = abs(Loc1[0] - Loc2[0]) +1
ydif = abs(Loc1[1] - Loc2[1]) +1
```

```python
def findareaBetweenCorners(L1, L2):
    xdif = abs(L1[0] - L2[0]) +1
    ydif = abs(L1[1] - L2[1]) +1
    multi = xdif * ydif
    return multi
```

```python
N = len(inputLocs)
maxArea = 0
for ii in range(N):
    Loc1 = inputLocs[ii]
    for jj in range(N):
        if ii!=jj:
            Loc2 = inputLocs[jj]
            miijj = findareaBetweenCorners(Loc1, Loc2)
            if miijj > maxArea:
                maxArea = miijj
            #print(f"for L1 = {Loc1}, L2 = {Loc2}, area is {miijj}")
print(maxArea)
```
