```python
def processMatrix(InMat):
    num_rows = len(InMat)
    num_cols = len(InMat[0])
    goodPaperRolls = 0
    #outMat = InMat # important for python learning... This is a reference to InMat
    outMat = [row[:] for row in InMat] # right way of copying the matrix

    for r, row in enumerate(InMat):
        for c, val in enumerate(row):
            if val!='.':
                if r == 0:
                    possibleRows = [r, r+1]
                elif r == (num_rows-1):
                    possibleRows = [r-1, r]
                else:
                    possibleRows = [r-1, r,  r+1]


                if c == 0:
                    possibleColumns = [c, c+1]
                elif c == (num_cols-1):
                    possibleColumns = [c-1, c]
                else:
                    possibleColumns = [c-1, c,  c+1]

                selected = []
                currentLocation = [r, c]
                for rx in possibleRows:
                    for cx in possibleColumns:
                        if [rx,cx] != currentLocation:
                            selected.append(InMat[rx][cx])

                count = selected.count("@")
                if count<4:
                    goodPaperRolls +=1
                    outMat[r][c] = "x"
                else:
                    outMat[r][c] = val

    return goodPaperRolls, outMat
```
