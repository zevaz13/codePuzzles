```python
def processInput(txtName):
        with open(txtName, 'r') as f:
                content = f.readlines()
        inputs = []
        outputs= []
        for line in content:
                colon_position = line.find(":")
                starter = line[:colon_position]
                inputs.append(starter)
                space_indexes = [i for i, char in enumerate(line) if char == ' ']

                groups = []
                for ti in space_indexes:
                        groups.append(line[ti+1:ti+4])
                outputs.append(groups)
        return inputs, outputs, { "inputs": inputs,"outputs": outputs}

def recPaths(IDX,OUPX,currentName):
        idx = IDX.index(currentName)
        inputs2test = OUPX[idx]
        total = 0
        for IdxName in inputs2test:
                if IdxName == "out":
                        total+=1
                elif IdxName == "you":
                        total+=0
                else:
                        total += recPaths(IDX, OUPX, IdxName)
        return total

def recPathsComps(IDX, OUPX, currentName):
    idx = IDX.index(currentName)
    inputs2test = OUPX[idx]

    paths = []

    for IdxName in inputs2test:
        if IdxName == "out":
            # Base case: full path is [currentName, "out"]
            paths.append([currentName, "out"])
        else:
            # Recursive case: get all subpaths
            subpaths = recPathsComps(IDX, OUPX, IdxName)

            # Prepend currentName to each returned path
            for sp in subpaths:
                paths.append([currentName] + sp)

    return paths

def recPathsCompsSum(IDX, OUPX, currentName, seen_fft=False, seen_dac=False):
    idx = IDX.index(currentName)
    inputs2test = OUPX[idx]
    print(currentName)
    # Update flags for this node
    if currentName == "fft":
        seen_fft = True
    if currentName == "dac":
        seen_dac = True

    total = 0

    for IdxName in inputs2test:
        if IdxName == "out":
            # Base case: check flags
            if seen_fft and seen_dac:
                total += 1
        else:
            # Recurse with updated flags
            total += recPathsCompsSum(IDX, OUPX, IdxName, seen_fft, seen_dac)

#     return total. Used AI for this one, my other solutions kept being too emmory intensive.
def recPathsCompsSumAI(
    IDX, OUPX, currentName,
    seen_fft=False, seen_dac=False,
    memo=None, visiting=None,
    name_to_idx=None
):
    # Initialize shared structures once
    if memo is None:
        memo = {}
    if visiting is None:
        visiting = set()
    if name_to_idx is None:
        name_to_idx = {name: i for i, name in enumerate(IDX)}

    # Cycle detection
    if currentName in visiting:
        return 0

    # Memoization: key is (node, seen_fft, seen_dac)
    key = (currentName, seen_fft, seen_dac)
    if key in memo:
        return memo[key]

    visiting.add(currentName)

    # Update flags
    if currentName == "fft":
        seen_fft = True
    if currentName == "dac":
        seen_dac = True

    idx = name_to_idx[currentName]
    children = OUPX[idx]

    total = 0

    for child in children:
        if child == "out":
            # Leaf: count if both flags seen
            if seen_fft and seen_dac:
                total += 1
        else:
            total += recPathsCompsSum(
                IDX, OUPX, child,
                seen_fft, seen_dac,
                memo, visiting,
                name_to_idx
            )

    visiting.remove(currentName)
    memo[key] = total
    return total
```
