```python
class CircleMover:
    def __init__(self, initial_loc, size=100):
        self.size = size
        self.current = initial_loc
        self.zero_crossings = 0

    def _split_distance(self, distance):
        hundreds = distance // self.size
        residual = distance % self.size
        return hundreds, residual
    def _move_left(self,distance, residual):
        # compute new location
        temp= (self.current - distance) % self.size
        # check if residual movement crosses zero
        if self.current-residual < 0:
            self.zero_crossings += 1
        return temp
    def _move_right(self,distance,residual):
        temp = (self.current + distance) % self.size

        # check if residual movement crossess zero
        if self.current + residual > self.size:
            self.zero_crossings +=1
        # special case: starting at zero and moving right. 
        if self.current == 0 and residual > 0:
            self.zero_crossings += 1
        return temp

    def move_sequence(self, tuple_inputs):
        for direction, distance in tuple_inputs:
            hundreds, residual = self._split_distance(distance)

            if direction == "L":
                temp = self._move_left(distance, residual)
            elif direction == "R":
                temp = self._move_right(distance, residual)
            else:
                raise ValueError(f"Invalid direction: {direction}")

            # add full hundreds of crossings
            self.zero_crossings += hundreds
            print(
                f"Move from {self.current}, dir={direction}, dist={distance}  "
                f"{temp}, hundreds={hundreds}, residual={residual}, "
                f"zeroCrossings={self.zero_crossings}"
            )
            self.current = temp
        return self.zero_crossings
```
