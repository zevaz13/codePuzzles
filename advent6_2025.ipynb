```python
from math import prod

def loadSetday6(fileName):
    with open(fileName,'r') as f:
        data = f.read()

    rows = [line.split() for line in data.strip().splitlines()]
    cols = list(zip(*rows))

    return cols
def flipStuff(cols):
    return 
def computeSumOfStuff(cols):
    sumOfStuff = 0
    for sy in cols:
        symbolStuff = sy[-1]
        nums = list(map(int, sy[:-1]))

        if symbolStuff == '+':
            sumOfStuff += sum(nums)
        if symbolStuff == '*':
            sumOfStuff += prod(nums)
    return(sumOfStuff)
```
