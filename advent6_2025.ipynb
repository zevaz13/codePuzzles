```python
import re
from math import prod  # make sure you have this

with open("day6test.txt") as f:
    lines = [line.rstrip("\n") for line in f]

# Find spans of non-space characters in the first non-empty line
first = next(line for line in lines if line.strip())
spans = [m.span() for m in re.finditer(r'\S+', first)]

# Extract columns (each span = one problem block)
columns = [
    [line[start:end] for line in lines[:-1]]
    for (start, end) in spans
]

# Your helper that gives you one string per problem; last char is the symbol
cols2flip = loadSetday6('day6test.txt')

symbols2use = [sy[-1] for sy in cols2flip]

numOps = len(symbols2use)

sumOfStuff = 0

for xi in range(numOps):
    aaa = columns[xi]          # list of rows (strings) for this problem
    currSym = symbols2use[xi]

    max_len2 = max(len(x) for x in aaa)

    # a[i] = list of chars in column i (topbottom)
    a = [
        [s[i] if i < len(s) else ' ' for s in aaa]
        for i in range(max_len2)
    ]

    # Part 2: read columns right-to-left, build numbers, skip empty ones
    anum = []
    for col in reversed(a):
        s = "".join(col).strip()
        if s:
            anum.append(int(s))

    if currSym == '+':
        sumOfStuff += sum(anum)
    elif currSym == '*':
        sumOfStuff += prod(anum)

print(sumOfStuff)
```

```python
from math import prod

def solve_part2(filename):
    with open(filename) as f:
        lines = [line.rstrip("\n") for line in f]

    # Separate number rows and operator row
    number_lines = lines[:-1]
    symbol_line  = lines[-1]

    height = len(number_lines)
    width  = max(len(line) for line in lines)

    # Pad all lines to same width
    number_lines = [line.ljust(width) for line in number_lines]
    symbol_line  = symbol_line.ljust(width)

    # A column is a separator if it's all spaces (numbers + symbol row)
    is_sep = [
        all(number_lines[r][c] == ' ' and symbol_line[c] == ' '
            for r in range(height))
        for c in range(width)
    ]

    # Group consecutive non-separator columns into problem blocks
    blocks = []
    in_block = False
    start = 0
    for c in range(width):
        if not is_sep[c] and not in_block:
            in_block = True
            start = c
        elif is_sep[c] and in_block:
            in_block = False
            blocks.append((start, c))
    if in_block:
        blocks.append((start, width))

    total = 0

    for start, end in blocks:
        # Operator: any non-space in symbol_line within this block
        op = next(ch for ch in symbol_line[start:end] if ch.strip())

        nums = []

        # Each column in [start, end) is a number, read right-to-left
        for c in range(end - 1, start - 1, -1):
            digits = ''.join(number_lines[r][c] for r in range(height))
            digits = digits.strip()
            if digits:
                nums.append(int(digits))

        if op == '+':
            total += sum(nums)
        else:
            total += prod(nums)

    return total


# Example:
print(solve_part2("day6test.txt"))
```
