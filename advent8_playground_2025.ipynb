```python
import numpy as np
data = np.loadtxt('day8_ex.txt',delimiter=',')
print(data)

N = len(data)
#D= np.zeros((N,N))
D = np.full((N, N), np.nan)
for ni in range(N):
    for di in range(N):
        #d = 0
        if  ni != di and di > ni:
            D[ni,di] = np.linalg.norm(data[ni,:]-data[di,:])
        #print(f"ni = {data[ni,:]}, di = {data[di,:]}, dist = {d}")
        #print(f"ni = {ni}, di = {di}")
#D = np.linalg.norm(data-data)
print(D)
```

```python
min_value = np.nanmin(D)
flat_min = np.nanargmin(D)
index_2d = np.unravel_index(flat_min, D.shape)
groups = []
for rep in range(10):
    pair = list(index_2d) 
    if not groups:
       groups.append(pair) 
    else:
        for gi in groups:
            flat = {x for p in gi for x in p}

            # Match if ANY element overlaps
            if pair[0] in flat or pair[1] in flat:
                gi.append(pair)
                placed = True
                break

        if not placed:
            groups.append([pair])
#if elements in list(index_2d) in gi, add them to gi.
            # Check all the elements, of groups. 
            # If list(index_2D) not in any of groups. start a new list and add it to groups 
print(groups)
```
