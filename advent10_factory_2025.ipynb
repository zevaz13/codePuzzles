```python
with open("day10_ex.txt", 'r') as f:
        content = f.readlines()
print(content)
```

```python
import re

def parse_line(s):
    # --- 1. Extract indicator block ---
    indicator_raw = re.search(r"\[(.*?)\]", s).group(1)
    indicator = [1 if c == "#" else 0 for c in indicator_raw]

    # --- 2. Extract all (...) groups ---
    groups_raw = re.findall(r"\((.*?)\)", s)
    groups = []
    for g in groups_raw:
        if g.strip() == "":
            groups.append([])
        else:
            groups.append([int(x) for x in g.split(",")])

    # --- 3. Extract joltage block ---
    joltage_raw = re.search(r"\{(.*?)\}", s).group(1)
    joltage = [int(x) for x in joltage_raw.split(",")]

    return indicator, groups, joltage
```

```python
line2use = content[2]

ind1, circuit1, jolt1 = parse_line(line2use)

print(f"for line {line2use} inds are {ind1}, wiring = {circuit1} and joltage is{jolt1}")
```
