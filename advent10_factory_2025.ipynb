```python
import re

def parse_line(s):
    # --- 1. Extract indicator block ---
    indicator_raw = re.search(r"\[(.*?)\]", s).group(1)
    indicator = [1 if c == "#" else 0 for c in indicator_raw]
    n_bits = len(indicator)

    # --- 2. Extract all (...) groups ---
    groups_raw = re.findall(r"\((.*?)\)", s)
    groups = []
    for g in groups_raw:
        if g.strip() == "":
            groups.append([])
        else:
            groups.append([int(x) for x in g.split(",")])

    # --- 2b. Convert groups to binary vectors ---
    binary_groups = []
    for g in groups:
        vec = [0] * n_bits
        for idx in g:
            vec[idx] = 1
        binary_groups.append(vec)

    # --- 3. Extract joltage block ---
    joltage_raw = re.search(r"\{(.*?)\}", s).group(1)
    joltage = [int(x) for x in joltage_raw.split(",")]

    # --- 4. Return dictionary ---
    return {
        "indicator": indicator,
        "groups": binary_groups,
        "num_groups": len(groups),
        "joltage": joltage
    }

def xor_bits(a, b):
    return [x ^ y for x, y in zip(a, b)]
```

```python
line2use = content[2]

ind1 = parse_line(line2use)

print(f"for line {line2use}, we have {ind1} ")
```

```python
import itertools
finalResults = []
for line in content:
    indx = parse_line(line)
    all_combs = []
    N = indx['num_groups']
    for i in range (1,N+1):
        all_combs.extend(itertools.combinations(range(N),i))

    #print(f"the combs are {all_combs}")
    numCombs = len(all_combs)
    #print(numCombs)

    for combi in all_combs:
        selected_groups = [indx['groups'][i] for i in combi]
        result = selected_groups[0]
        if len(selected_groups) > 1:
                for g in selected_groups[1:]:
                    result = xor_bits(result, g)
        
        iterationx = xor_bits(indx['indicator'],result)
        if sum(iterationx) == 0:
            numVals = len(selected_groups)
            break

    print(numVals)
    finalResults.append(numVals)


print(f"Final result is {sum(finalResults)}")
```

```python
from functools import cache
from itertools import combinations, product

def build_pattern_table(button_vectors: list[tuple[int, ...]]) -> dict[tuple[int, ...], dict[tuple[int, ...], int]]:
    num_buttons = len(button_vectors)
    num_dims = len(button_vectors[0])

    pattern_table = {
        parity: {} 
        for parity in product(range(2), repeat=num_dims)
    }

    for k in range(num_buttons + 1):
        for chosen in combinations(range(num_buttons), k):
            summed = tuple(
                map(sum, zip((0,) * num_dims, *(button_vectors[i] for i in chosen)))
            )
            parity = tuple(x % 2 for x in summed)

            if summed not in pattern_table[parity]:
                pattern_table[parity][summed] = k

    return pattern_table


def solve_target(button_vectors: list[tuple[int, ...]], target: tuple[int, ...]) -> int:
    pattern_table = build_pattern_table(button_vectors)

    @cache
    def recurse(goal: tuple[int, ...]) -> int:
        if all(x == 0 for x in goal):
            return 0

        best = 10**9
        parity = tuple(x % 2 for x in goal)

        for pattern, cost in pattern_table[parity].items():
            if all(p <= g for p, g in zip(pattern, goal)):
                new_goal = tuple((g - p) // 2 for p, g in zip(pattern, goal))
                best = min(best, cost + 2 * recurse(new_goal))

        return best

    return recurse(target)


def solve_input(raw: str):
    total = 0
    lines = raw.splitlines()

    for idx, line in enumerate(lines, 1):
        _, *vecs, tgt = line.split()

        target = tuple(int(x) for x in tgt[1:-1].split(","))
        vecs = [[int(x) for x in v[1:-1].split(",")] for v in vecs]

        # Convert to indicator vectors
        button_vectors = [
            tuple(int(dim in row) for dim in range(len(target)))
            for row in vecs
        ]

        score = solve_target(button_vectors, target)
        total += score

    print(total)
```

```python

with open("day10_tst.txt", 'r') as f:
        content = f.read()
solve_input(content)
```
