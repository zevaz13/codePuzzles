```python
def findIdx(s, ch): # This function returns all the occurences of a pattern on a line. Returns a list.
    return [i for i, ltr in enumerate(s) if ltr == ch]

with open("day7_test.txt",'r') as f:
    data = f.readlines() # load the input file as a list of lines of strings

#print(data)

beamIndex = findIdx(data[0],"S")  # initial location for the beam
countStuff = 0                    # start the count for the beam deviations

for di in range(1,len(data)): # for each line of the input, starting from the second, and leaving the last one out (since only dots.)

    beamLocs = findIdx(data[di],"^") # find Beam deviators
    common_items_set = set(beamIndex) & set(beamLocs) # compare the location of beam deviators in present line with the currelnt beams

    #print(f"Line {di}: {data[di]} has {beamLocs}. Current locations are {beamIndex}")
    if not common_items_set: # if there are no beam deviators, just keep them as they are 
        print("No beam thing")
    else:
        countStuff += len(common_items_set) # if there exist beam deviators, add one to the count.

        for li in common_items_set:  #for each beam deviator that interacts with the current beams. Remove the current, add the one before and after.
            #print(f"common val {li}")
            beamIndex.remove(li)
            beamIndex.append(li-1)
            beamIndex.append(li+1)
    beamIndex = list(set(beamIndex)) # there are some deviators that produce the same beam twice. This line removes those.
   
print(countStuff)
```

```python
def quantum_tachyon(grid, currentRow, currentColumn):

    # grid refers to the whole input. Think about how to simplify 
    # currentRow. Where are we in traversing the grid row-wise
    # currentColumn. Where is the current beam located. 
    totalRows    = len(grid)    
    totalColumns = len(grid[0])
    
    #print(f"CRow = {currentRow}, Ccol = {currentColumn}")
    if currentRow == (totalRows - 1): # We have completed a whole grid 
        return 1

    if grid[currentRow + 1][currentColumn] ==".":
        return  quantum_tachyon(grid, currentRow+1, currentColumn)

    if grid[currentRow +1][currentColumn] =="^":
        total_paths  = 0 
        total_paths+=quantum_tachyon(grid, currentRow+1, currentColumn+1)
        total_paths+=quantum_tachyon(grid, currentRow+1, currentColumn-1)
        return total_paths

    return 0
    
```
