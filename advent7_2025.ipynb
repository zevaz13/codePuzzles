```python
def findIdx(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]

with open("day7_test.txt",'r') as f:
    data = f.readlines()

print(data)

beamIndex = findIdx(data[0],"S")  # initial location for the beam
countStuff = 0

for di in range(1,len(data)):

    beamLocs = findIdx(data[di],"^")
    common_items_set = set(beamIndex) & set(beamLocs)

    print(f"Line {di}: {data[di]} has {beamLocs}. Current locations are {beamIndex}")
    if not common_items_set:
        print("No beam thing")
    else:
        countStuff += len(common_items_set)

        for li in common_items_set: 
            print(f"common val {li}")
            beamIndex.remove(li)
            beamIndex.append(li-1)
            beamIndex.append(li+1)
    beamIndex = list(set(beamIndex))
   
print(countStuff)
```
