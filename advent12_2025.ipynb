```python
sizesArrays = [7, 7, 6, 7, 5, 7]
maxSizes    = [9, 9, 9, 9, 9, 9]
with open("day12_test.txt", 'r') as f:
        content = f.readlines()
```

```python


Line0 = content[0]
colon = Line0.find(":")
print(Line0.find(":"))
W = int(Line0[0:Line0.find("x")])
H = int(Line0[Line0.find("x")+1:colon])

nums = [int(x) for x in Line0[colon+1:].split()]
result = [a * b for a, b in zip(sizesArrays, nums)]
resultMax = [a * b for a, b in zip(maxSizes, nums)]
maxArea = W*H
```

```python
sumSizes = 0
sumAreas = 0

for line in content:
    colon = line.find(":")
    W = int(line[0:line.find("x")])
    H = int(line[line.find("x")+1:colon])
    maxArea = W*H

    nums = [int(x) for x in line[colon+1:].split()]
    result = [a * b for a, b in zip(sizesArrays, nums)]
    resultMax = [a * b for a, b in zip(maxSizes, nums)]

    if sum(result) <= maxArea:
        sumSizes += 1

    if sum(resultMax) <= maxArea:
        sumAreas += 1
```

```python
print(sumSizes)
print(sumAreas)
```
